#**DESAFIOS AULA 02**

##Desafio 01: Ordenar o gráfico countplot

```python
cod_compostos = dados['composto'].value_counts().sort_values(ascending=False).index[:5]
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=dados.query('composto in @cod_compostos'), x='composto', 
                   order=cod_compostos)
ax.set_title("Top 5 compostos")
ax.set_xlabel('Compostos');
```

##Desafio 02: Melhorar a visualização alterando o tamanho da fonte.


```python
cod_compostos = dados.query('composto in @cod_compostos')
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(
    x = 'composto', 
    data=cod_compostos, 
    order=cod_compostos['composto'].value_counts(ascending=True).index)
ax.set_title('Top 5 compostos', fontweight='bold', fontsize=16)
ax.set_xlabel('Composto', fontweight='bold')
ax.set_ylabel('Contagem', fontweight='bold')
for p in ax.patches:
  ax.annotate(p.get_height(), (p.get_x()+0.3, p.get_height()))            # O laço for com o annotate escreve o valor de cada barra acima dela
plt.show()
```
