##Desafio 06: Renomear as colunas tirando o h√≠fen.

```python
nome_colunas = dados.columns
novo_nome = []

for i in nome_colunas:
    novo_nome.append(i.replace('-',''))
        
dados.columns = novo_nome
dados.head()
```

##Desafio 07: Deixar os gr√°ficos bonit√µes (Matplotlib.pyplot)

```python
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
plt.figure(figsize=(8,8))
sns.histplot(data=dados, x='g0', kde=True)
plt.title("Distribui√ß√£o da express√£o gen√©tica do G0", fontsize=20);
```

```python
plt.figure(figsize=(8,8))
plt.hist(x='g-1', data=dados, bins=20)
plt.title('Distribui√ß√£o da Express√£o gen√©tica do G1', fontsize=20);
```
